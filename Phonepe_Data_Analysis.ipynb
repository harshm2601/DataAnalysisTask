```python
#importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Data
#Data Source :- https://www.kaggle.com/datasets/komalkhetlani/upi-payment-apps-review-google-play-store/data

google_data = pd.read_csv("GooglePayIndia.csv")
paytm_data = pd.read_csv("PaytmIndia.csv")
phonepay_data = pd.read_csv("PhonePayIndia.csv")
```

```python
google_data = google_data[google_data["thumbsUpCount"] <=500]
paytm_data = paytm_data[paytm_data["thumbsUpCount"] <=550]
phonepay_data = phonepay_data[phonepay_data["thumbsUpCount"] <=100]
```

```python
google_data.drop(columns= ["Unnamed: 0","reviewId" , "userName", "userImage" ,"content" ,"reviewCreatedVersion" , "at" ,"replyContent" ,"repliedAt"] ,axis=1 , inplace=True)
google_data.info()
```

```python
phonepay_data.drop(columns= ["Unnamed: 0","reviewId" , "userName", "userImage" ,"content" ,"reviewCreatedVersion" , "at" ,"replyContent" ,"repliedAt"] ,axis=1 , inplace=True)
phonepay_data.info()
```

```python
paytm_data.drop(columns= ["Unnamed: 0","reviewId" , "userName", "userImage" ,"content" ,"reviewCreatedVersion" , "at" ,"replyContent" ,"repliedAt"] ,axis=1 , inplace=True)
paytm_data.info()
```

```python
google_data["upi_company"] = "Google"
paytm_data["upi_company"] = "paytm"
phonepay_data["upi_company"] = "phonepay"
```

```python
google_data = google_data.sample(frac=1).reset_index(drop=True)
paytm_data = paytm_data.sample(frac=1).reset_index(drop=True)
phonepay_data = phonepay_data.sample(frac=1).reset_index(drop=True)

data = pd.concat([phonepay_data ,paytm_data[:11735 ], google_data[:11735]], ignore_index=True)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="score", y="thumbsUpCount", hue="upi_company", data=data)
plt.title('Bar Plot To Show which Company has more number of votes in all scores Category')
plt.show()
```

implementation process.

<b>Features and Functionality:</b>

<ul>
<li>Single QR code accepts payments from UPI, credit/debit cards, and wallets </li>
<li>Real-time settlements for merchants </li>
<li>Transaction history and data insights for merchants </li>
</ul>

<b>Implementation Process: </b>
<ul>
<li>Merchants register on PhonePe app and generate a unique QR code</li>
<li>Customers scan the QR code using their PhonePe app and enter payment amount</li>
<li>Secure and instant payment confirmation for both parties </li>
</ul>
<b>Benefits for Merchants: </b>
<ul>
<li>Simplified payment acceptance process (one QR for all) </li>
<li>Faster settlements lead to improved cash flow </li>
<li>Access to valuable business data for informed decisions </li>
<li>Potential for promotional campaigns through PhonePe platform </li>
</ul>

<b>Benefits for Customers: </b>
<ul>
<li>Seamless and quick payment experience</li>
<li>Wide acceptance across merchants </li>
<li>Security and convenience of using a trusted platform</li>
</ul>

```python
data1 = pd.read_csv('Data_Aggregated_Transaction_Table.csv')
data2 = pd.read_csv('Data_Aggregated_User_Summary_Table.csv')
data3 = pd.read_csv('Data_Aggregated_User_Table.csv')
data4 = pd.read_csv('Data_Map_Districts_Longitude_Latitude.csv')
data5 = pd.read_csv('Data_Map_IndiaStates_TU.csv')
data6 = pd.read_csv('Data_Map_Transaction_Table.csv')
data7 = pd.read_csv('Data_Map_User_Table.csv')

```

```python
print(data1.info())
print(data2.info())
print(data3.info())
print(data4.info())
print(data5.info())
print(data6.info())
print(data7.info())
```

```python
grouped_data = data1.groupby(['Payment_Mode', 'Year'])['Total_Transactions_count'].sum().unstack()
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Payment Mode')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions by Payment Mode and Year')
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

As we can clearly see in above graph , transactions has risen exponentially from year 2018 to 2022 ,with one of supporting Factor being "all-in-one Qr Strategy"

```python
plt.figure(figsize=(6, 6))
plt.pie(data1.groupby('Payment_Mode')['Total_Transactions_count'].sum(), labels=data1['Payment_Mode'].unique(), autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Transactions by Payment Mode')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Given Chart shows that most of payments made are either peer-peer or financial services Transactions.

Phone pe Interface and "all in one" strategy is specifically designed for such transactions
